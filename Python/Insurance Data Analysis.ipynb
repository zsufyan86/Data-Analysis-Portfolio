# Goal:

The goal of this analysis is to gain insights from the insurance dataset to help in creating a model that predicts the cost of medical insurance based on various input features.

#### Step 1: Import Libraries and Load Dataset
1. Import the necessary libraries such as Pandas, Matplotlib, NumPy, and Seaborn.
2. Load the insurance dataset using Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load the dataset
insurance_df = pd.read_csv("C:\Data Analytics - Master course Simplilearn\Course 3 - Programming Basics and Data Analytics with Python\insurance.csv")
```

#### Step 2: Check Shape and Data Types
1. Check the shape of the dataset (number of rows and columns).
2. Check the data types of each column.


```python
# Check the shape of the data
print("Shape of the dataset:", insurance_df.shape)

# Check the data types of each column
print(insurance_df.dtypes)
```

#### Step 3: Check Missing Values
1. Check for missing values in the dataset.
2. Determine appropriate measures to handle missing values.


```python
# Check for missing values
print("Missing values:\n", insurance_df.isnull().sum())
```

#### Step 4: Explore Relationship Between Features and Target
1. Use count plots for categorical columns.
2. Use scatter plots for numerical columns to explore the relationship between features and the target column (charges).


```python
# Count plot of categorical columns
sns.countplot(x='smoker', data=insurance_df)
plt.title('Count of Smokers vs. Non-Smokers')
plt.show()

# Scatter plot of numerical columns (e.g., age vs charges)
sns.scatterplot(x='age', y='charges', data=insurance_df)
plt.title('Scatter Plot of Age vs. Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

```

#### Step 5: Data Visualization
1. Use various plots (histograms, box plots, etc.) to visualize the relationship between different features.


```python
# Histogram of BMI distribution
sns.histplot(data=insurance_df, x="bmi", bins=30, kde=True)
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()

# Box plot of charges vs. smoker status
sns.boxplot(x='smoker', y='charges', data=insurance_df)
plt.title("Charges vs. Smoker Status")
plt.xlabel("Smoker")
plt.ylabel("Charges")
plt.show()
```

```python
# Pairplot of numerical columns
sns.pairplot(insurance_df)
plt.show()
```

#### Step 6: Analysis of Premium Charges for Smokers vs. Non-Smokers with Age
1. Use line plots or scatter plots to analyze how the number of premium charges varies with age for smokers and non-smokers.


```python
# Create a line plot to visualize the trend of charges vs age for smokers and non-smokers
sns.lineplot(x='age', y='charges', hue='smoker', data=insurance_df)
plt.title('Trend of Charges vs. Age for Smokers and Non-Smokers')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

**Observations:**

- The dataset contains information about insurance policyholders including features like age, sex, BMI, children, smoker status, region, and charges.
- There are no missing values in the dataset.
- The count plot shows that there are more non-smokers than smokers in the dataset.
- The scatter plot of age vs charges shows a positive correlation between age and charges.
- The pairplot provides a visual overview of the relationships between different numerical columns.
- The line plot indicates that premium charges tend to increase with age for both smokers and non-smokers, but the increase is more significant for smokers.
